```python
# a
freq_a = 12
deltat_a = 1/freq_a
T_a = (2000-1990)*freq_a
print("--- a ---\nf:",freq_a,"dt:",deltat_a,"T:",T_a,"\n")
# 
# b
freq_b = 1
deltat_b = 1/freq_b
T_b = (2011-2000)*freq_b
print("--- b ---\nf:",freq_b,"dt:",deltat_b,"T:",T_b,"\n")

# c
freq_c = 365 # assuming 365 days a year
T_c = 4*freq_c
deltat_c = 1/freq_c
print("--- c ---\nf:",freq_c,"dt:",deltat_c,"T:",T_c,"\n")

# d
freq_d = 252 # assuming 252 trading days a year
deltat_d = 1/freq_d
T_d = (2010-2000)*freq_d # Example
print("--- d ---\nf:",freq_d,"dt:",deltat_d,"T:",T_d,"\n")

# e 
freq_e = 7 # measurements per week
deltat_e = 1/freq_e
T_e = 4*freq_e
print("--- e ---\nf:",freq_e,"dt:",deltat_e,"T:",T_e,"\n")

# f
freq_f = 7
deltat_f = 1/freq_f
T_f = 4*52*freq_f # approximately
print("--- f ---\nf:",freq_f,"dt:",deltat_f,"T:",T_f,"\n")
```

```python
# a
import pandas as pd
dd = pd.read_csv("http://stat.ethz.ch/Teaching/Datasets/WBL/rainDay.txt",sep=" ")
dd.head()

print(dd["DATE"])
dd["DATE"] = pd.to_datetime(dd["DATE"],format="%d.%m.%Y")
print(dd["DATE"])

# b --> Redundant
```

```python
# c
import matplotlib.pyplot as plt
plt.plot(dd["DATE"],dd["rain"])
plt.xlabel("Date")
plt.ylabel("Rainfall [mm]")
plt.title("Daily Rainfall (01.01.2000-31.12.2008)")
plt.show()

dd_clipped = dd[(dd["DATE"] >= "01.01.2006") & (dd["DATE"] <= "01.01.2007")]
plt.plot(dd_clipped["DATE"],dd_clipped["rain"])
plt.xlabel("Date")
plt.ylabel("Rainfall [mm]")
plt.title("Daily Rainfall (01.01.2006-01.01.2007)")
plt.show()
```
