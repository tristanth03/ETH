# Stationary?

```r
plot(AirPassengers)
```

**NOT** stationary since it clearly has a strong trend

```r
plot(lynx)
```

**NOT** stationary since it has periodical oscillations

```r
plot(lh)
```

Most likely stationary, no clear sign of rejection.

```r
plot(EuStockMarkets[, "SMI"])
```

**NOT** stationary, it has stochastic seasonality and a trend.

```r
plot(ts(log(EuStockMarkets[, "SMI"][2:length(EuStockMarkets[, "SMI"])]/EuStockMarkets[, "SMI"][1:length(EuStockMarkets[, "SMI"])-1])))
```

Most likely stationary, there is no sign of trend, seasonality, major variant shifts, and ultimately it is centered around a constant
