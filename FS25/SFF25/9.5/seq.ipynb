```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def as_convergence(n):
    x = np.random.uniform(0,1,n)
    X = ((x>=0)&(x<=1/n))
    return np.mean(X)

 
n,m = 28,20
for i in range(n):
    s = 0
    I = int(10**(i/4))
    for j in range(m):
        s+=as_convergence(I)

    plt.plot(I,s/(m),'o')
plt.xscale('log')
plt.show()
```
