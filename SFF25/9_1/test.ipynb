```python
import numpy as np
import matplotlib.pyplot as plt
n = 100000
x = np.random.uniform(0,1,n)
y = -np.log(x)
z = np.random.exponential(1,n)
fig, axs = plt.subplots(3,1,figsize=(14,14))
axs[0].hist(x,density=True)
axs[1].hist(y,density=True)
axs[2].hist(z,density=True)
plt.show()
```

```python
X = np.random.binomial(1,0.5,size=(10000,500))
Y = []
for i in range(500):
    y = 0
    for j in range(10000):
       
        y += (X[j,i]*0.5**(j+1))

    Y.append(y)

plt.hist(Y)
plt.show()
```

```python
n = 1000
Mu = np.random.standard_normal(n)**2
S = np.random.standard_normal(n)**2+1
D = np.random.uniform(0+0.001,1-0.001,10)
def f(x,mu,s_sq,delta):
    return 1/(np.abs(x)**delta) * 1/(np.sqrt(2*np.pi*s_sq)) * np.exp(-0.5* (x-mu)**2/s_sq)

fl = []
fr = []
from math import gamma
X = np.linspace(-10_000,10_000,10_000)
for i in range(n):
    mu = Mu[i]
    s_sq = S[i]
    x = np.random.normal(mu,s_sq,n)
    for j in range(len(D)):
        delta = D[j]

        theta = ((1-delta)/2)
        fr.append(2**(theta)*gamma(theta) * 1/(s_sq**delta *np.sqrt(2*np.pi)))
        fl.append(sum(f(X,s_sq,mu,delta)))

R = []
for i in range(len(fl)):
    for j in range(len(fl)):
        r = (fl[i]-fr[i])
        R.append((r<=0))

print(R)
```
