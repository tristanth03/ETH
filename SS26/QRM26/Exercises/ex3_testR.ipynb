```r
GDAXI <- read.csv("GDAXI.csv")
GDAXI <- GDAXI[nrow(GDAXI):1, ] 
GDAXI <- ts(GDAXI[,5])

GSPC <- read.csv("GSPC.csv")
GSPC <- GSPC[nrow(GSPC):1, ] 
GSPC <- ts(GSPC[,5])
```

```r
plot(GDAXI)
plot(GSPC)
```

```r
plot(diff(GDAXI))
plot(diff(GSPC))
```

```r
lr_GDAXI<- diff(log(GDAXI))
lr_GSPC<- diff(log(GSPC))
pacf(lr_GDAXI)
pacf(lr_GSPC)
```
