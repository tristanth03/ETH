```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def val(S_T):
    phi = np.array([(S_T<k1),-(S_T<k2),(S_T>=k3),-(S_T>=k4),-((S_T<k1)+-(S_T<k2)+(S_T>=k3)+(S_T>=k4))]).astype(int)
    S = np.array([k1,k2,k3,k4,S_T])
    return np.transpose(phi)@S


k1,k2,k3,k4 = 80,90,110,120


S_v = [x for x in range(k4+10)]
l = []
for s in S_v:
    l.append(val(s))
plt.plot(S_v,l,'.')

```
