```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
def s_path(sigma,r,T,d):

    S = np.zeros(shape=(T,d))
    W = np.zeros(shape=(T,d))
    S[0,:] = 1
    W[0,:] = 0
    for t in range(1,T):
        W[t,:] = W[t-1,:]+np.sqrt(t)*np.random.normal(loc=0,scale=1,size=d)
        S[t,:] = S[t-1,:]*np.exp((r-0.5*sigma**2)*t+sigma*W[t,:])

    return W,S
```

```python
W,S = s_path(sigma=0.1e-3,r=-0.04e-3,T=100,d=1000)
```

```python
plt.plot(W)
plt.show()
plt.plot(S)
plt.show()
```

```python
def turbo_binary(S,K,j,p):
    return np.power(S,p)*(j*S > j*K)
```

```python
K = 1; j = -1; p = -1
plt.plot(S[-1,:],turbo_binary(S[-1,:],K,j,p),'.')
plt.axhline(y=K,color='red',label="K")
plt.legend()
plt.show()
```
