```python
import pandas as pd
import calendar
import numpy as np
```

```python
dd = pd.read_csv("milkData.csv")

print(dd["time(milk)"])
toDate = lambda decimal : (calendar.monthrange(int(decimal),round((decimal-int(decimal))*12+1))[1])
dd["days"]=dd["time(milk)"].apply(toDate)
dd["milk (30d adj)"]=dd["milk"]*30/dd["days"]
dd.head()
import matplotlib.pyplot as plt
plt.plot(dd["time(milk)"],dd["milk (30d adj)"],label="Adj")
plt.plot(dd["time(milk)"],dd["milk"])
plt.legend
plt.show()

# calendar.monthrange(1994,1)
```
