```r
# a
freq_a = 12
deltat_a = 1/freq_a
T_a = (2000-1990)*freq_a
cat("--- a ---\nf:",freq_a,"dt:",deltat_a,"T:",T_a,"\n")
# 
# b
freq_b = 1
deltat_b = 1/freq_b
T_b = (2011-2000)*freq_b
cat("--- b ---\nf:",freq_b,"dt:",deltat_b,"T:",T_b,"\n")

# c
freq_c = 365 # assuming 365 days a year
T_c = 4*freq_c
deltat_c = 1/freq_c
cat("--- c ---\nf:",freq_c,"dt:",deltat_c,"T:",T_c,"\n")

# d
freq_d = 252 # assuming 252 trading days a year
deltat_d = 1/freq_d
T_d = (2010-2000)*freq_d # Example
cat("--- d ---\nf:",freq_d,"dt:",deltat_d,"T:",T_d,"\n")

# e 
freq_e = 7 # measurements per week
deltat_e = 1/freq_e
T_e = 4*freq_e
cat("--- e ---\nf:",freq_e,"dt:",deltat_e,"T:",T_e,"\n")

# f
freq_f = 7
deltat_f = 1/freq_f
T_f = 4*52*freq_f # approximately
cat("--- f ---\nf:",freq_f,"dt:",deltat_f,"T:",T_f,"\n")
```

```r
# a
dd <- read.table("http://stat.ethz.ch/Teaching/Datasets/WBL/rainDay.txt", header=T);
head(dd)
```

```r
# b
dd[["DATE"]] <- as.Date(dd[["DATE"]], format = "%d.%m.%Y")
f = 365
d_ts <- ts(dd[["rain"]],frequency = f,start = c(2000,1))
head(dd)
```

```r
# c
plot(d_ts,xlab = "Date",ylab="Rainfall [mm]",main="Daily Rainfall (01.01.2000-31.12.2008)")
d_ts_clipped <- window(x = d_ts,start = c(2006,1),end=c(2007,1))
plot(d_ts_clipped,xlab = "Date",ylab="Rainfall [mm]",main="Daily Rainfall (01.01.2006-01.01.2007)")
```
